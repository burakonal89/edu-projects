```python
import pandas as pd
import numpy as np
import datetime
import re
pd.set_option("display.max_colwidth", 1000)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def convert_names(name, names_dict):
    return names_dict[name]
```

```python
names_dict = {"Sabri Uğur Sönmez": "Ugur",
              "Halit Sinan Yıldırım": "Sinan",
              "Halil Burak Oz": "HBO",
              "Caglar Keskin": "Caglar",
              "Burak": "BO",
              "Gökhan Bestel": "Zeki",
              "Yigit Aykan": "Yigit",
              "Erdem Kulahci": "Erdem",
              "Seyda Can Yılmaz": "Seyda",
              "Mustafa Yahsi": "Yahsi",
              "Ömer Özkiper": "Omer",
              "Emre İşeri": "Emre",
              "Mehmet Ihsan Orgut": "Mehmet",
              "Alp Özgen": "Alp",
              "Emre Iseri": "Emre"}
def pre_process(f_in, f_out):
    df = pd.DataFrame(columns=["date", "time", "name", "message"])
    info_pattern = re.compile(r'(\d\d/\d\d/\d\d\d\d),\s(\d\d:\d\d:\d\d):\s')
    iter_ = 0
    with open(f_in, "r") as f_in, open(f_out, "w") as f_out:
        for line in f_in:
            line = line.strip()
            info = info_pattern.match(line)
            if info:
                date = info.group(1)
                time = info.group(2)
                body = line[21:]
                if ":" not in body:
                    name = "info_message"
                    message = body.strip()
                else:
                    name, message = body.split(":", 1)
                    name = name.strip()
                    try:
                        name = convert_names(name, names_dict)
                        message = message.strip()
                # print iter_, message
                        text = date+" "+time+"\t"+name+"\t"+message+"\n"
                        f_out.write(text)
                    except Exception, e:
                        print e
                        print name
                        continue
```

```python
pre_process("_chat.txt", "chat_bros.csv")
# pre_process("_chat_p.txt", "chat_pijler.csv")
```

```python
df = pd.read_csv("chat_bros.csv", 
                 sep="\t", 
                 names=["date-time", "name", "message"])
# converting date-time to date time object
df['date-time'] = df['date-time'].apply(lambda x: datetime.datetime.strptime(x, "%d/%m/%Y %H:%M:%S"))
# changing unknown numbers or duplicates
df["name"] = df["name"].apply(lambda x: "Erdem Kulahci" if "829" in x else x)
```

```python
df["name"] = df["name"].apply(lambda x: "Emre İşeri" if "‪019‬" in x else x)
```

```python
df_groupped_day = df.set_index('date-time').groupby(pd.TimeGrouper("D")).count()
df_groupped_week = df.set_index('date-time').groupby(pd.TimeGrouper("W")).count()
df_groupped_month = df.set_index('date-time').groupby(pd.TimeGrouper("M")).count()
```

```python
plt.plot_date(df_groupped_day.index, 
              df_groupped_day["message"], 
              fmt="-")
plt.title("Daily # of messages")
```

```python
N = 14 # 2 weeks window
plt.plot_date(df_groupped_day.index, 
              np.convolve(df_groupped_day["message"], np.ones((N,))/N, mode='same'), 
              fmt="-")
plt.title("Daily # of messages - Moving Average - 14 days")
df_groupped_day.iloc[np.argmax(np.convolve(df_groupped_day["message"], np.ones((N,))/N, mode='same'))]
```

```python
plt.plot_date(df_groupped_week.index, 
              df_groupped_week["message"], 
              fmt="-")
plt.title("Weekly # of messages")
```

```python
plt.plot_date(df_groupped_month.index, 
              df_groupped_month["message"], 
              fmt="-")
plt.title("Monthly # of messages")
```
